# Web Scraping

Web Scraping of the Mamaearth beauty products website has been performed using Selenium. Data points like Product Name, Product Link, Product Price, Product Rating, Product Reviews, Product Offers and Main Ingredient have been scraped.
This data is then put in a dataframe, and exported as an Excel sheet.

```python
import selenium
from selenium import webdriver as wb

webd = wb.Chrome("C:/Users/tulik/Desktop/chromedriver.exe")
webd.get('https://mamaearth.in/product-category/beauty')
```

```python
#scraping the entire section containing all products
ele1=webd.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/section/section')
```

```python
ele1.text
```

```python
#common class uniquewhite contains all products links and details
productinfolist=webd.find_elements_by_class_name("uniquewhite")
```

```python
#scraping the links from all a tags using href property
linkslist=[]

for prod in productinfolist:
    pp2=prod.find_element_by_tag_name('a')
    linkslist.append(pp2.get_property('href'))
    
#pp=productinfolist[0]
```

```python
linkslist
```

```python
from tqdm import tqdm
from selenium.common.exceptions import NoSuchElementException
allproducts=[]

#scraping individual data points from all products using XPath
for i in tqdm(linkslist):
    webd.get(i)
    try:
        productName=webd.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[2]/div[2]/div/h1/div').text #name
    except NoSuchElementException:
        productName=''
    try:
        productMRP=webd.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[2]/div[2]/div/div[2]/div[1]').text #price
    except NoSuchElementException:
        productMRP=''  
    try:   
        productRating=webd.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[2]/div[2]/div/div[1]/a[1]/div/div/div[1]').text #rating
    except NoSuchElementException:
        productRating=''  
    try:
        productReviews=webd.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[2]/div[2]/div/div[1]/a[2]').text #reviews
    except NoSuchElementException:
        productReviews=''
    try:    
        productOffers=webd.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[2]/div[2]/div/div[2]/div[3]').text #offers
    except NoSuchElementException:
        productOffers=''
    try:    
        productIngredient=webd.find_element_by_xpath('//*[@id="__next"]/div[2]/div[1]/div[3]/div[2]/div/ul/li[1]/span/strong').text #ingredient
    except NoSuchElementException:
        productIngredient='' 
        
    temp={'productName':productName, 
      'productLink':i,
     'productMRP':productMRP,
     'productRating':productRating,
     'productReviews':productReviews,
     'productOffers':productOffers,
     'productIngredient':productIngredient}
    allproducts.append(temp)
webd = wb.Chrome("C:/Users/tulik/Desktop/chromedriver.exe")
webd.get(linkslist[0])
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(allproducts)
```
